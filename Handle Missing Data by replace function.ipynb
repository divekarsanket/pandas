# Handling Missing Data - replace method

```python
import pandas as pd
import numpy as np
df = pd.read_csv("wh_data.csv")
df
```

# Replacing single value

```python
new_df = df.replace(-99999, value=np.NaN)
new_df
```

```python
new_df = df.replace([-99999, -88888, '0'], np.NaN)
new_df
```

# Replacing per column

```python
new_df = df.replace({
        'temperature': -99999,
        'windspeed': -99999,
        'event': '0'
    }, np.nan)
new_df
```
