```python
import pandas as pd
```

```python
df = pd.read_csv('wheather_data.csv')
```

```python
df
```

```python
df.set_index('day',inplace=True)
df
```

```python
new_df = df.fillna(0)
new_df
#cover NaN with 0
```

```python
new_df = df.fillna({
    'temperature':0,
    'windspeed':0,
    'event':'no_event'
})
new_df
#alternative method can be assign separate values to columns one by one
```

```python
new_df = df.fillna(method="ffill")
new_df
#carry forward last row value to next row and fills empty values
#can also use bfill - only difference is just take next row value
```

```python
new_df = df.interpolate()
new_df
#for interpolation
```

```python
#to drop row with NaN value
new_df = df.dropna()
new_df
```
